```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('final.csv')
df.columns
```

```python
data = df.groupby(['AdminYear', 'DISTRICT']).size().reset_index(name='students_count')
data.head(15)
```

```python
pivot_data = data.pivot_table(index='AdminYear', columns=['DISTRICT'], values='students_count', fill_value=0)
```

```python
pivot_data.head()
```

```python
# Prepare the SARIMA model and make predictions
predictions = {}
for district in pivot_data.columns:
    ts = pivot_data[district].dropna()
    if len(ts) > 1:  # Ensure there are enough data points to train the model
        try:
            model = SARIMAX(ts, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
            model_fit = model.fit(disp=False)
            forecast = model_fit.forecast(steps=1)
            if len(forecast) > 0:  # Ensure forecast has values
                predictions[district] = forecast[0]
            else:
                print(f"No forecast values for {district}")
        except Exception as e:
            print(f"Error in modeling {district}: {str(e)}")
```

```python
predictions_df = pd.DataFrame.from_dict(predictions, orient='index', columns=['predicted_students_enrolled'])
predictions_df.index.name = 'DISTRICT'
predictions_df.head()
```

```python
# Optionally, plot the predictions for each district
for district in predictions_df.index:
    ts = pivot_data[district].dropna()
    plt.figure(figsize=(10, 6))
    plt.plot(ts.index, ts, label='Actual Enrollments')
    if district in predictions:
        plt.plot(ts.index[-1] + 1, predictions[district], 'ro', label='Predicted Enrollment')
    plt.title(f'Enrollments for District: {district}')
    plt.xlabel('Year')
    plt.ylabel('Number of Students Enrolled')
    plt.legend()
    plt.show()
```
