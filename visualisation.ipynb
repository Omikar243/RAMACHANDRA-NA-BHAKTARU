```python
import plotly.express as px
# Combine all data for the plot
combined_data = df[df['AdminYear'].isin(years)]

# Set up colorscale for years
custom_colors = px.colors.qualitative.Set3[:len(years)]

# Create the Plotly figure using px.histogram with a grouped bar mode
fig = px.histogram(combined_data, x='DISTRICT', color='AdminYear',
                   color_discrete_sequence=custom_colors,
                   title='Count of Districts by Year',
                   labels={'DISTRICT': 'District', 'AdminYear': 'Years', 'count': 'Count'},
                   barmode='group')

# Update layout for better readability and interactivity
fig.update_layout(
    xaxis={'title': {'text': 'District', 'font': {'size': 14, 'family': 'Arial, sans-serif'}}, 'categoryorder': 'total descending'},
    yaxis={'title': {'text': 'Count', 'font': {'size': 14, 'family': 'Arial, sans-serif'}}},
    legend={'title': {'text': 'Years', 'font': {'size': 12, 'family': 'Arial, sans-serif'}}},
    font={'family': 'Arial, sans-serif', 'size': 12},
    title={'text': 'Count of Districts by Year', 'font': {'size': 18, 'family': 'Arial, sans-serif'}},
)
fig.update_layout(height=1100, width=1400) 
# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
# Assuming 'df' is your DataFrame
dataframne=df.copy()
# Set the AdminYear as the index for time series analysis
dataframne.set_index('AdminYear', inplace=True)

# Plot time series for categorical fields DISTRICT
fields_to_plot = ['DISTRICT']

fig = go.Figure()
first_category = True
for field in fields_to_plot:
    # Group by AdminYear and field, and count the occurrences
    grouped_data = dataframne.groupby(['AdminYear', field]).size().unstack(fill_value=0)
    # Add a trace for each category in the field
    for category in grouped_data.columns:
        visible = True if first_category else 'legendonly'
        first_category = False
        fig.add_trace(go.Scatter(
            x=grouped_data.index,
            y=grouped_data[category],
            mode='lines+markers',
            name=f'{field}: {category}',
            visible=visible
        ))

# Update the layout
fig.update_layout(
    title='Time Series of DISTRICT counts over years',
    xaxis_title='Year',
    yaxis_title='Count of DISTRICT',
    legend_title='Categories',
    height=1100,
    width=1400
)
# Show the plot
fig.show()
```

```python
# Create a dropdown menu for selecting a district
dropdown_buttons = []

# List of unique districts
districts = df['DISTRICT'].unique()

# Create the figure
fig = go.Figure()

# Add all traces and set visibility to False initially
for district in districts:
    filtered_df = df[df['DISTRICT'] == district]
    grouped_data = filtered_df.groupby(['AdminYear', 'PINCODE']).size().unstack(fill_value=0)
    for pincode in grouped_data.columns:
        fig.add_trace(go.Scatter(
            x=grouped_data.index,
            y=grouped_data[pincode],
            mode='lines+markers',
            name=f'{district}: {pincode}',
            visible=False  # Set to False initially
        ))

# Create a dropdown menu for selecting a district
for district in districts:
    # Create a list of visibility states
    visibility = [trace.name.startswith(district) for trace in fig.data]
    # Add the button for the dropdown
    dropdown_buttons.append(dict(
        label=district,
        method='update',
        args=[{'visible': visibility},
              {'title': f'Time Series of PINCODE counts over years for {district}'}]
    ))
# Update the layout with dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.75,
            xanchor="left",
            y=1.25,
            yanchor="top"
        ),
    ],
    title='Time Series of PINCODE counts over years',
    xaxis_title='Year',
    yaxis_title='Count of PINCODE',
    legend_title='Categories',
    height=1100,
    width=1400
)
# Show the plot
fig.show()
```

```python
# Create a dropdown menu for selecting a district
district_dropdown_buttons = []

# List of unique districts
districts = df['DISTRICT'].unique()

# Create the figure
fig = go.Figure()

# Add all traces and set visibility to False initially
for district in districts:
    filtered_df = df[df['DISTRICT'] == district]
    grouped_data = filtered_df.groupby(['AdminYear', 'PINCODE']).size().unstack(fill_value=0)
    for i, pincode in enumerate(grouped_data.columns):
        fig.add_trace(go.Bar(
            x=grouped_data.index,
            y=grouped_data[pincode],
            name=f'{district}: {pincode}',
            visible=False if i != 0 else True  # Set only the first pincode of each district to visible initially
        ))

# Create a dropdown menu for selecting a district
for district in districts:
    # Create a list of visibility states for each district
    visibility = []
    for trace in fig.data:
        if trace.name.startswith(district):
            visibility.append(True)
            break  # Show only the first pincode initially
        else:
            visibility.append(False)
    
    # Add the button for the dropdown
    district_dropdown_buttons.append(dict(
        label=district,
        method='update',
        args=[{'visible': visibility},
              {'title': f'Time Series of PINCODE counts over years for {district}'}]
    ))

# Create a dropdown for pin codes for detailed visibility control
pin_code_dropdown_buttons = []
for district in districts:
    filtered_df = df[df['DISTRICT'] == district]
    grouped_data = filtered_df.groupby(['AdminYear', 'PINCODE']).size().unstack(fill_value=0)
    pin_codes = grouped_data.columns
    
    for pin_code in pin_codes:
        pin_code_visibility = [trace.name == f'{district}: {pin_code}' for trace in fig.data]
        pin_code_dropdown_buttons.append(dict(
            label=f'{district}: {pin_code}',
            method='update',
            args=[{'visible': pin_code_visibility},
                  {'title': f'Time Series of PINCODE counts over years for {district}: {pin_code}'}]
        ))
# Update the layout with district dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=district_dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.60,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
        dict(
            buttons=pin_code_dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.77,
            xanchor="left",
            y=1.15,
            yanchor="top"
        )
    ],
    title='Time Series of PINCODE counts over years',
    xaxis_title='Year',
    yaxis_title='Count of PINCODE',
    legend_title='Categories',
    height=800, width=1400
)
# Show the plot
fig.show()
```
